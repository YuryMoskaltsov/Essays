```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
cd /Users/yurymoskaltsov/Documents/Data_Science/Datasets/oil_prices
```

```python
df_oil = pd.read_excel("Oil_Prices_From_2000.xlsx", parse_dates = ["Date"], index_col = "Date")
```

```python
print(df_oil.head(5))
```

```python
df_oil.rename(columns ={"Closing Value": "Price"}, inplace = True)
print(df_oil.columns)
```

```python
print(max(df_oil.Price))
```

```python
df_oil = df_oil.resample("Y").mean()
x_oil = df_oil.index
y_oil = df_oil.Price

plt.figure(figsize = (20,8))
plt.xlabel("Year")
plt.ylabel("Price Per Barrel")
plt.plot(x_oil, y_oil, color = "black")


```

```python
df_gini = pd.read_excel("Gini_coefficient_world.xls")
print(df_gini.head(5))
```

```python
df_gini = df_gini.fillna(0)
df_gini_russia = df_gini[df_gini["Country Code"] == "RUS"]
print(df_gini_russia.head(5))
```

```python
df_gini_russia_edited = df_gini_russia.drop(["Country Name", "Country Code","Indicator Name", "Indicator Code"], axis = 1)
years = df_gini_russia_edited.columns
years = np.array(years)
years = years.reshape(-1,1)
print(years.shape)
values = np.array(df_gini_russia_edited)
values = values.reshape(-1,1)
print(values.shape)
```

```python
df_gini_russia_years = pd.DataFrame(years, columns = ["year"]).reset_index()
df_gini_russia_values = pd.DataFrame(values, columns = ["gini_coefficient"]).reset_index()
df_gini_russia_final = pd.merge(df_gini_russia_years,df_gini_russia_values)
```

```python
x_gini = df_gini_russia_final.year.iloc[40:]
y_gini = df_gini_russia_final.gini_coefficient.iloc[40:]


plt.figure(figsize= (20,8))
plt.grid(color = "grey", linewidth = 1, axis = "y")
plt.ylim(35,50)
plt.title("Gini Coefficient Russia", fontsize = 35, weight = "bold")
plt.xlabel("Year", fontsize = 30)
plt.ylabel("Coefficient Number", fontsize = 30)
plt.plot(x_gini, y_gini, color = "orange")





```

```python
df_putin_support = pd.read_excel("Putin_Support.xlsx")
print(df_putin_support)
```

```python
df_putin_support = df_putin_support.sort_values('Year', ascending = True)
print(df_putin_support.Year[0:10])
```

```python
x_support_putin = df_putin_support.Year
y_support_putin = df_putin_support.Support
```

```python
plt.figure(figsize = (20,8))
plt.ylabel("Support %", fontsize = 30)
plt.xlabel("Year", fontsize = 30)
plt.title("Putin Support", fontsize = 40, fontweight =20)
plt.plot(x_support_putin, y_support_putin,color = "red")
plt.show()
```
